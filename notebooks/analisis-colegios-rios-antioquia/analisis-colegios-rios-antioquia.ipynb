```python
import geopandas as gpd
```

```python
departamentos = gpd.read_file('./data/Departamentos.shp')
```

```python
departamentos.head()
```

```python
departamentos.plot()
```

```python
rios = gpd.read_file('./data/Drenajedoble.shp')
```

```python
rios.head()
```

```python
colegios = gpd.read_file('./data/Establecimientoseducativos.shp')
```

```python
colegios.head()
```

```python
antioquia = departamentos[departamentos['DeNombre'] == 'Antioquia']
```

```python
antioquia.plot()
```

```python
rios_antioquia = gpd.overlay(antioquia, rios, how = "intersection")
```

```python
rios_antioquia.head()
```

```python
colegios_antioquia = colegios.query("DeNombre == 'ANTIOQUIA'") 
colegios_antioquia.head()
```

```python
colegios['DeNombre'].unique()
```

```python
axis = antioquia.plot(figsize = (20,12), alpha = 0.3)
rios_antioquia.plot( ax = axis, color = 'blue')
colegios_antioquia.plot(ax = axis, color = 'green')
```

```python
rios_antioquia['buffer']= rios_antioquia.buffer(500)
```

```python
rios_antioquia['buffer'].plot()
```

```python
rios_antioquia = rios_antioquia.set_geometry("buffer")
```

```python
colegios_afectados = gpd.overlay( colegios_antioquia, rios_antioquia, how = "intersection")
```

```python
colegios_afectados.plot()
```

```python
axis = antioquia.plot(figsize = (20,12), alpha = 0.3)
rios_antioquia.plot( ax = axis, color = 'blue')
colegios_afectados.plot(ax = axis, color = 'green')
```

```python
axis  = colegios_antioquia.plot(figsize = (20, 12))
colegios_afectados.plot(ax = axis, color = 'red')
```
